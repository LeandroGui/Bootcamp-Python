```python
# Ejercicio 2: POO

class Bicicleteria:
    def __init__(self, bicicletas=[], ganancias=0, cantidad_de_ventas=0):
        self.bicicletas = bicicletas
        self.ganancias = ganancias
        self.cantidad_de_ventas = cantidad_de_ventas

    def vender_bicicleta(self,bicicleta):
        print(f"Esta por vender la bicicleta con nº de serie {bicicleta[0]}!!!")
        for i in self.bicicletas:
            if i == bicicleta:
                self.ganancias += bicicleta[1]
                self.cantidad_de_ventas += 1
                self.bicicletas.remove(i)

    def comprar_bicicleta(self, bicicleta):
        self.bicicletas.append(bicicleta)


    class Bicicleta():
        def __init__(self, nro_de_serie="", modelo="", anio=0, precio=0):
            self.__nro_de_serie = nro_de_serie
            self.__modelo = modelo
            self.__anio = anio
            self.__precio = precio

        def set_precio(self, precio):
            self.__precio = precio
            return f"Se modificó correctamente el precio de la bicicleta!!!"
        def get_precio(self):
            return self.__precio
        def get_nro_de_serie(self):
            return self.__nro_de_serie

# Consignas del examen resueltas:

# Creamos un objeto bicicleteria
bicicleteria_1 = Bicicleteria(bicicletas=[], ganancias=0, cantidad_de_ventas=0)

# Creamos 4 objetos bicicleta
bicicleta_1 = Bicicleteria.Bicicleta("AAA0001","Calle",2015,50000.51)
bicicleta_2 = Bicicleteria.Bicicleta("AAA0002","Playera",2012,35000)
bicicleta_3 = Bicicleteria.Bicicleta("AAA0003","Cross",2018,75354.70)
bicicleta_4 = Bicicleteria.Bicicleta("AAA0004","Mountain",2021,95635)

# Caratula Sistema
print("====================================================================================")
print("                                   SISTEMA BICICLETERIA                             ")
print("                                    By Leandro Guiñazú                              ")
print("====================================================================================")

# Utilizamos metodos get y set de la Clase "Bicicleta" en objeto "bicicleta_1"
print("*******************MÉTODOS DE VISUALIZACIÓN Y MODIFICACIÓN**************************")
print("Precio de la bicicleta 1: $",bicicleta_1.get_precio())
print("Modificando precio bicicleta 1...",bicicleta_1.set_precio(52000))
print("Nuevo precio bicicleta 1: $",bicicleta_1.get_precio())
print("Visualizamos nº de serie de bicicleta 1: ",bicicleta_1.get_nro_de_serie())

# Utilizamos metodo "comprar_bicicleta()" de la Clase "Bicicleteria" en objeto "bicicleteria_1"
print("************************************************************************************")
print("************************************INVENTARIO**************************************")
print("Cargando bicicletas compradas al inventario...")
bicicleteria_1.comprar_bicicleta([bicicleta_1.get_nro_de_serie(),bicicleta_1.get_precio()])
bicicleteria_1.comprar_bicicleta([bicicleta_2.get_nro_de_serie(),bicicleta_2.get_precio()])
bicicleteria_1.comprar_bicicleta([bicicleta_3.get_nro_de_serie(),bicicleta_3.get_precio()])
bicicleteria_1.comprar_bicicleta([bicicleta_4.get_nro_de_serie(),bicicleta_4.get_precio()])
print("Lista de bicicletas por Nº de serie y precio: ", bicicleteria_1.bicicletas)

# Utilizamos metodo "vender_bicicleta()" de la Clase "Bicicleteria" en objetos "bicicleta_4" y "bicicleta_2"
print("************************************************************************************")
print("************************************VENTAS******************************************")
bicicleteria_1.vender_bicicleta([bicicleta_4.get_nro_de_serie(),bicicleta_4.get_precio()])
print("Lista de bicicletas por Nº de serie y precio: ", bicicleteria_1.bicicletas)
bicicleteria_1.vender_bicicleta([bicicleta_2.get_nro_de_serie(),bicicleta_2.get_precio()])
print("Lista de bicicletas por Nº de serie y precio: ", bicicleteria_1.bicicletas)

print("************************************************************************************")
print("************************************UTILIDADES**************************************")
print("Ganancia total: $",bicicleteria_1.ganancias)
print("Cantidad de ventas totales: ",bicicleteria_1.cantidad_de_ventas)
```

```python
# 4. ¿Cuanto dinero se recaudo en ese viaje?
recaudacion_titanic = df_titanic['Fare'].sum()
print(f"Se recaudo $ {recaudacion_titanic}")
```

```python
# 5. ¿Cuál es la cantidad de personas que viajaban en primera clase?

primera_clase = df_titanic.apply(lambda x: x['Pclass'] == 1, axis=1).sum()
print(f"En primera clase viajaban {primera_clase} personas.")
```

```python
# 6. Graficar la cantidad de personas que viajaban por cada clase del barco.

plt.ylabel('Cantidad de Personas')
plt.xlabel('Clase')
plt.title('Cantidad de personas que viajaban por cada clase')
df_titanic['Pclass'].value_counts().plot(kind='bar')
```
