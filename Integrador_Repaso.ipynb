```python
# EJERCICIO 2: Programación Orientada a Objetos

# La Fundación Santísima Trinidad está interesada en construir un software para
# manejar los alumnos y sus notas. Es por ello que contrato a un programador de
# Python para realizar dicho proyecto.

# ATRIBUTOS

# ● matricula: número entero de matrícula
# ● nombre: nombre del estudiante
# ● apellido: apellido del estudiante
# ● fecha_nacimiento: la fecha de nacimiento debe tener el siguiente
# formato: [DD-MM-YYYY] por ejemplo 23-01-2002. A partir de este
# atributo podremos calcular la edad. La fecha debe tener el formato
# datetime. (VER COMO TRANSFORMAR UN STRING A DATETIME)
# ● exámenes: un diccionario con la fecha del examen (en formato
# datetime) y la nota del estudiante.
# ● notafinal: es el promedio de las notas.

# MÉTODOS

# ● __str__(): devuelve el nombre y apellido del alumno, la matrícula y la
# nota final.
# ● __getitem__(fecha): devuelve la nota de una fecha de examen.
# ● __setitem__(fecha, nota): modifica la nota de una fecha de examen.
# ● listarExamenes(): devuelve las fechas con las notas que obtuvo el
# alumno. Al final devolverá la nota final promedio.
# ● getEdad(): devuelve la edad de acuerdo a la fecha de nacimiento del
# estudiante. (INVESTIGAR COMO SACAR LA EDAD DE UNA FECHA
# CON DATETIME)
# ● filtroFechas(fecha_desde): mostrar las fechas de exámenes tomados
# después de cierta fecha.
# ● perfilEstudiante(): mostrar el número de matrícula, nombre, apellido,
# edad, y listar los exámenes del alumno (reutilizar método).

# Una vez construida la clase, vamos a construir un software que pueda
# gestionar estos estudiantes. Para ello vamos a utilizar funciones!
# Debemos programar las siguientes funciones:

# ● cargar_examen(): Esta función cargará la nota, y la fecha (en formato
# datetime) a un diccionario. Debemos poder ingresar todos exámenes
# como el usuario quiera. Debe retornar un diccionario con todos los
# exámenes.
# ● registrarEstudiante(listaEstudiantes): Esta función debe solicitar los
# datos del constructor de la clase Estudiantes. Crear el objeto y agregarlo
# a la lista estudiantes, una lista con todos los objetos estudiantes
# creados.
# ● listadoEstudiantes(listaEstudiantes): Deberá hacer un print de todos
# los objetos de la lista. (Haciendo uso del método mágico __str__)
# ● buscarEstudiante(listaEstudiantes): Esta función deberá solicitar la
# matricula del objeto estudiante que queremos buscar. Para ello debemos
# recorrer la listaEstudiantes en búsqueda de aquel objeto que tenga el
# número de matrícula que coincida con la búsqueda. Si la encuentra
# deberá retornar el objeto, sino mostrar que no se encontró y retornar
# None.

# ● modificarNotas(listaEstudiantes): Debemos buscar al estudiante que
# queremos modificar la nota (reutilizar función). Si no se encontró el
# objeto la función retorna None, pero si lo encuentra, pedir la fecha de
# examen que queremos modificar la nota. Hacemos uso del método
# mágico __get__ del objeto Estudiante, si no lo encuentra indicar que no
# se encontró esa fecha de examen. Luego preguntar al usuario sea cual
# sea el caso (que exista o no esa fecha), si se desea agregar/modificar la
# nota. En caso afirmativo, modificar la nota con la nueva nota solicitada al
# usuario (haciendo uso del método mágico __set__).
# ● listarMatriculas(listaEstudiantes): Recorrer la lista de estudiantes y
# agregar a otra lista las matriculas de cada uno de esos objetos. Al final
# retornar la lista de matrículas y mostrarlas de manera formateada (una
# debajo de otra por ejemplo).
# ● filtrarExamenesEstudiante(listaEstudiantes): Buscar un estudiante
# especifico, solicitar la fecha desde y ejecutar el método filtroFechas() del
# objeto pasándole la fecha desde.
# ● listarNotasEstudiante(listaEstudiantes): Buscar un estudiante
# particular y ejecutar el método del objeto listarExamenes().
# ● fichaEstudiante(listaEstudiantes): Buscar un estudiante y ejecutar el
# método del objeto perfilEstudiante().
# ● main(listaEstudiantes): Esta función será la función principal que
# muestra el menú de opciones de las funciones anteriores, y recibe una
# opción. Luego de acuerdo a la opción elegida se ejecutara una de esas
# funciones. El programa comenzará cuando se llame este método.
# RECORDAR INFORMAR AL USUARIO EL ESTADO DEL SISTEMA EN
# TODO MOMENTO POR EJEMPLO:
# Si accedo a la función modificar notas, informar que estamos dentro de esa
# función con un título. Y si modificamos algo informar con un mensaje, por
# ejemplo:

# Nota actualizada!
# presione enter para continuar….

from datetime import datetime
from dateutil.relativedelta import relativedelta

class Estudiantes():

    #Constructor de la clase donde los datos pueden estar vacios.
    def __init__(self, matricula, nombre, apellido, fecha_nacimiento, examenes):#, nota_final):


        #Atributos de la instancia
        self.matricula = matricula      #Privado
        self.nombre = nombre  #Privado
        self.apellido = apellido
        self.fecha_nacimiento = datetime.strptime(fecha_nacimiento, '%d-%m-%Y') #DATETIME
        self.examenes = examenes
        self.nota_final = sum(examenes.values()) / len(examenes.values())

    #Metodos

    def __str__(self):
        return "No. Matricula: {} - {} {} - Nota Final: {}".format(self.matricula, self.nombre, self.apellido, self.nota_final)

    def __getitem__(self, fecha):
        convertir_fecha = datetime.strptime(fecha, '%d-%m-%Y')
        return self.examenes.get(convertir_fecha, "No se encontro la fecha indicada")

    def __setitem__(self, fecha, nota):
        convertir_fecha = datetime.strptime(fecha, '%d-%m-%Y')
        self.examenes[convertir_fecha] = nota

    def listar_Examenes(self):
        for fecha, nota in self.examenes.items():
            print("Fecha: ", str(fecha).split()[0], "\nNota: ", nota, "\n\n")
        print("Nota final: ", self.nota_final)

    def get_Edad(self):
        edad = relativedelta(datetime.now(), self.fecha_nacimiento)
        return edad.years

    def filtro_fechas(self, fecha_desde):
        print("Fechas de examenes despues de ", fecha_desde+":")
        convertir_fecha = datetime.strptime(fecha_desde, '%d-%m-%Y')
        for fecha in self.examenes.keys():
            if fecha >= convertir_fecha:
                print(str(fecha).split()[0])

    def perfil_estudiante(self):
        print("Nro. Matricula {}\nNombre: {}\nApellido: {}\nEdad: {}".format(self.matricula, self.nombre, self.apellido, self.get_Edad()))
        self.listar_Examenes()

# Funciones Programa

def cargar_examen():
    print("Cargar examen\n")
    examen = {}
    rpta = True
    while rpta:
        nota = int(input("Ingrese Nota: "))
        fecha = datetime.strptime(input("Ingrese fecha del examen [dd-mm-yyyy]: "), '%d-%m-%Y')
        examen[fecha] = nota
        rpta = input("Desea agregar otro examen?").upper() in ("S","Si")

    return examen

def registrar_estudiante(lista_estudiantes):
    #lista_estudiantes = []
    print("Registro de Estudiantes\n")
    matricula = int(input("Nº de matricula: "))
    nombre = input("Nombre: ")
    apellido = input("Apellido: ")
    fecha_nacimiento = input("Fecha de nacimiento [dd-mm-yyyy]: ")
    exam = cargar_examen()
    objeto_alumno = Estudiantes(matricula, nombre, apellido, fecha_nacimiento, exam)
    lista_estudiantes.append(objeto_alumno)

def listado_estudiantes(lista_estudiantes):
    print("Listado de Estudiantes\n")
    for objeto_alumno in lista_estudiantes:
        print(objeto_alumno)

def buscar_estudiante(lista_estudiantes):
    print("Buscando Estudiante.......\n")
    matricula = int(input("Nº matricula a buscar: "))
    for objeto_alumno in lista_estudiantes:
        if matricula == objeto_alumno.matricula:
            print("Alumno encontrado!!!")
            print(objeto_alumno)
            return objeto_alumno

    print("No se encontro el alumno buscado")


def modificar_notas(lista_estudiantes):
    print("Modificar Notas\n")
    objeto_alumno = buscar_estudiante(lista_estudiantes)
    if objeto_alumno == None:
        return
    else:
        fecha_examen = input("Fecha del examen [dd-mm-yyyy]: ")
        print(objeto_alumno[fecha_examen])
        rpta = input("Desea agregar/modificar la nota? [S/Si]: ").upper() in ("S","Si")
        if rpta:
            nueva_nota = int(input("Ingrese nueva nota: "))
            objeto_alumno[fecha_examen] = nueva_nota
            input("...NOTA ACTUALIZADA... \nPresione enter para continuar")

def listar_matriculas(lista_estudiantes):
    print("Listar matriculas")
    lista_matriculas = []
    for objeto_alumno in lista_estudiantes:
        lista_matriculas.append(objeto_alumno.matricula)

    return lista_matriculas

def filtrar_examen_estudiante(lista_estudiantes):
    print("Filtrar notas de estudiantes")
    objeto_alumno = buscar_estudiante(lista_estudiantes)
    fecha_desde = input("Fecha desde donde quiere filtrar examenes [dd-mm-yyyy]: ")
    objeto_alumno.filtro_fechas(fecha_desde)

def listar_nota_estudiante(lista_estudiantes):
    print("Listar notas de estudiantes")
    objeto_alumno = buscar_estudiante(lista_estudiantes)
    objeto_alumno.listar_Examenes()

def ficha_estudiante(lista_estudiantes):
    print("Ficha del Estudiante")
    objeto_alumno = buscar_estudiante(lista_estudiantes)
    objeto_alumno.perfil_estudiante()

def main(lista_estudiantes):   # Funcion principal (menu).
    opc_menu = ''
    while opc_menu != '9':
        print("==============================")
        print("      By Leandro Guiñazu      ")
        print("      GitHub:                 ")
        print("==============================")
        print("\n")
        print("|****************************|")
        print("|**|         MENU         |**|")
        print("|**|      DE OPCIONES     |**|")
        print("|****************************|")
        print("")

        if opc_menu == '1':
            registrar_estudiante(lista_estudiantes)
        elif opc_menu == '2':
            listado_estudiantes(lista_estudiantes)
        elif opc_menu == '3':
            buscar_estudiante(lista_estudiantes)
        elif opc_menu == '4':
            modificar_notas(lista_estudiantes)
        elif opc_menu == '5':
            matriculas = listar_matriculas(lista_estudiantes)
            for matricula in matriculas:
                print(matricula)
        elif opc_menu == '6':
            filtrar_examen_estudiante(lista_estudiantes)
        elif opc_menu == '7':
            listar_nota_estudiante(lista_estudiantes)
        elif opc_menu == '8':
            ficha_estudiante(lista_estudiantes)
        opc_menu = (input('Menú de opciones:\n(1) Registrar Estudiante\n(2) Listado Estudiantes\n(3) Buscar Estudiante\n(4) Modificar Notas\n(5) Listar Matriculas\n(6) Filtrar fechas de examenes de un estudiante\n(7) Mostrar notas de estudiante\n(8) Mostrar perfil del estudiante\n(9) Terminar\nElige una opción: '))


lista_estudiantes = []
main(lista_estudiantes)
```
